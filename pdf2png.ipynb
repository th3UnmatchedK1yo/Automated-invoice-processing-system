```python
import os
from pdf2image import convert_from_path

PDF_FOLDER = "/mnt/c/Users/Legion/Documents/jimmy tran/Automated-invoice-processing-system/dataset/pdf_invoices"
OUTPUT_FOLDER = "/mnt/c/Users/Legion/Documents/jimmy tran/Automated-invoice-processing-system/dataset/converted_pngs"
os.makedirs(OUTPUT_FOLDER, exist_ok=True)

for filename in os.listdir(PDF_FOLDER):
    if filename.lower().endswith(".pdf"):
        pdf_path = os.path.join(PDF_FOLDER, filename)
        base_name = os.path.splitext(filename)[0]

        print(f"ğŸ“„ Converting: {filename} â†’ {base_name}.png")

        try:
            # Chá»‰ láº¥y trang Ä‘áº§u (náº¿u cháº¯c cháº¯n má»—i hÃ³a Ä‘Æ¡n chá»‰ cÃ³ 1 trang)
            images = convert_from_path(pdf_path, dpi=300, first_page=1, last_page=1)
            img = images[0]
            output_path = os.path.join(OUTPUT_FOLDER, f"{base_name}.png")
            img.save(output_path, "PNG")

            print(f"âœ… Done: {output_path}")
        except Exception as e:
            print(f"âŒ Lá»—i vá»›i {filename}: {e}")
```
