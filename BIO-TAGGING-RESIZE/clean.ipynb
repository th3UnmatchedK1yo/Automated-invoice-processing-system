```python
# ── cell / file: resize_bboxes.py ────────────────────────────────────────────
from pathlib import Path
import json

# ─────────────────── CONFIG ────────────────────────────────────────────────
SRC_DIR  = Path(r"/mnt/c/Users/Legion/Documents/jimmy tran/Automated-invoice-processing-system/BIO-TAGGING-RENAME/qhuy_data")     # <── folder with id-fixed files
DST_DIR  = Path(r"/mnt/c/Users/Legion/Documents/jimmy tran/Automated-invoice-processing-system/BIO-TAGGING-RESIZE/qhuy_data")   # <── will be created if needed
IMG_W, IMG_H = 2481, 3508                          # original image size in pixels

DST_DIR.mkdir(exist_ok=True)

def norm_bbox(b):
    x0, y0, x1, y1 = b
    return [
        int(1000 * x0 / IMG_W),
        int(1000 * y0 / IMG_H),
        int(1000 * x1 / IMG_W),
        int(1000 * y1 / IMG_H),
    ]

for p in SRC_DIR.glob("*.json"):
    data = json.loads(p.read_text(encoding="utf-8"))
    data["bboxes"] = [norm_bbox(b) for b in data["bboxes"]]

    out = DST_DIR / p.name
    out.write_text(json.dumps(data, ensure_ascii=False, indent=2), encoding="utf-8")
    print("✓ bbox-norm:", p.name)

print(f"Done.  Files written to {DST_DIR}")
```
