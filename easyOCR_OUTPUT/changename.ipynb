```python
import os
import json

# === Paths ===
jsonl_path = "/mnt/c/Users/Legion/Documents/jimmy tran/Automated-invoice-processing-system/data/labeled_invoices_fixed.jsonl"
folder_path = "/mnt/c/Users/Legion/Documents/jimmy tran/Automated-invoice-processing-system/easyOCR_OUTPUT/output_rename"

# === Read all new IDs from the JSONL file ===
with open(jsonl_path, "r", encoding="utf-8") as f:
    ids = [json.loads(line)["id"] for line in f if line.strip()]

# === List and sort old files (assumes .json extension) ===
old_files = sorted([f for f in os.listdir(folder_path) if f.endswith(".json")])

# === Safety check ===
if len(old_files) != len(ids):
    print(f"❌ Number of files ({len(old_files)}) does not match number of IDs ({len(ids)}). Aborting.")
else:
    for old_name, new_id in zip(old_files, ids):
        old_path = os.path.join(folder_path, old_name)
        new_path = os.path.join(folder_path, f"{new_id}.json")
        os.rename(old_path, new_path)
        print(f"✅ Renamed: {old_name} → {new_id}.json")
```
