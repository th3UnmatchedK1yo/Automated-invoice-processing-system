```python
# ── cell / file: json_pretty_to_singleline.py ───────────────────────────────
from pathlib import Path
import json       # <-- separate line
import shutil     # <-- if you actually need shutil later


# ─────────────────── CONFIG ────────────────────────────────────────────────
SRC_DIR = Path(r"/mnt/c/Users/Legion/Documents/jimmy tran/Automated-invoice-processing-system/BIO-TAGGING-RETAG/qhuy_data")          # retagged, pretty-printed JSONs
DST_DIR = Path(r"/mnt/c/Users/Legion/Documents/jimmy tran/Automated-invoice-processing-system/BIO-TAGGING-SingleLINE/qhuy_data")         # will be created
DST_DIR.mkdir(exist_ok=True)

for p in SRC_DIR.glob("*.json"):
    data = json.loads(p.read_text(encoding="utf-8"))

    # dump with no spaces → separators=(",", ":")
    single_line = json.dumps(data, ensure_ascii=False, separators=(",", ":"))

    (DST_DIR / p.name).write_text(single_line + "\n", encoding="utf-8")
    print("✓ single-line:", p.name)

print(f"Done.  Files written to {DST_DIR}")
```
